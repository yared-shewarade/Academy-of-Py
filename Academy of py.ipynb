# School Summary

Create an overview table that summarizes key metrics about each school, including:

    -> School Name
    -> School Type
    -> Total Students
    -> Total School Budget
    -> Per Student Budget
    -> Average Math Score
    -> Average Reading Score
    -> % Passing Math
    -> % Passing Reading
    -> Overall Passing Rate (Average of the above two)
    -> Create a dataframe to hold the above results

```python
# determining school type and school name
School_types = school_df.set_index(["school_name"])["type"]
```

```python
# Total Students per school
TotalStudent_PerSchool = SchoolData["school_name"].value_counts()
```

```python
# Total School Budget
Budget_PerSchool = SchoolData.groupby(["school_name"]).mean()["budget"]
```

```python
# Total budget per students
Budget_PerStudent = Budget_PerSchool/ TotalStudent_PerSchool
```

```python
# Average MAth and Reading Score
AvgMath_Score = SchoolData.groupby(["school_name"]).mean()["math_score"]
AvgReading_Score = SchoolData.groupby(["school_name"]).mean()["reading_score"]
```

```python
PassMath_Score =  SchoolData[SchoolData["math_score"] > 70].groupby("school_name").count()["student_name"]
PassReading_Score =  SchoolData[SchoolData["reading_score"] > 70].groupby("school_name").count()["student_name"]
```

```python
# Average percenting rate for math and reading
PassMath_Percent = PassMath_Score / TotalStudent_PerSchool * 100
PassReading_Percent = PassReading_Score / TotalStudent_PerSchool * 100
```

```python
# Overall Passing Rate (Average of the above two)
OverallPassing_Rate = (PassMath_Percent + PassReading_Percent) / 2
```

```python
# Create a dataframe to hold the above results
School_Summary = pd.DataFrame({"School Type":School_types, 
                              "Total Students":TotalStudent_PerSchool,
                              "Total School Budget":Budget_PerSchool, 
                              "Per Student Budget":Budget_PerStudent,
                              "Average Math Score":AvgMath_Score, 
                              "Average Reading Score":AvgReading_Score,
                              "% Passing Math":PassMath_Percent, 
                              "% Passing Reading":PassReading_Percent,
                              "% Overall Passing Rate":OverallPassing_Rate})
```

```python
School_Summary
```

Top Performing Schools (By Passing Rate)

```python
# decending sorting to get top five schools in overall passing rate
Top_Schools = School_Summary.sort_values(["% Overall Passing Rate"], ascending=False)
```

# Scores by School Spending

Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

    -> Average Math Score
    -> Average Reading Score
    -> % Passing Math
    -> % Passing Reading
    -> Overall Passing Rate (Average of the above two)

```python
# Average Spending Ranges pers students
Budget_PerStudent = Budget_PerSchool /TotalStudent_PerSchool
```

```python
bins = [0,585,615,645,675]
labels =["<$585","$585-615","$615-645","$645-675"]
```

```python
School_Summary["Spending Ranges"] = pd.cut(Budget_PerStudent, bins, labels=labels) 
```

```python
# Calculating the average spending range for MAth and Reading per student
SpendMath = School_Summary.groupby("Spending Ranges").mean()["Average Math Score"]
SpendReading = School_Summary.groupby("Spending Ranges").mean()["Average Reading Score"]
SpendMathPass = School_Summary.groupby("Spending Ranges").mean()["% Passing Math"]
SpendReadPass = School_Summary.groupby("Spending Ranges").mean()["% Passing Reading"]
SpendOverallPass = (SpendMath + SpendReading)/2
```

```python
# Creating pnadas dataframe from spending summary
School_Summary_BySpending = pd.DataFrame({"Average Math Score":SpendMath,
                                          "Avereage Reading Score":SpendReading,
                                          "% Passing Math":SpendMathPass,
                                          "% Passing Reading":SpendReadPass,
                                          "% Overall Passing Rate":SpendOverallPass})
```

```python
School_Summary_BySpending
```
